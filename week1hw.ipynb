```python
import numpy as np
import matplotlib.pyplot as plt

canvas = np.zeros((16, 16), dtype=int)

canvas[4:10, 4] = 1
canvas[4:10, 8] = 1
canvas[4, 4:9] = 1
canvas[6, 4:8] = 1

canvas[0, :] = 1
canvas[-1, :] = 1
canvas[:, 0] = 1
canvas[:, -1] = 1

plt.imshow(canvas, cmap='gray')
plt.axis('off')
plt.show()
```
